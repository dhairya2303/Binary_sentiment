```python
train_data = train_data.dropna()
train_data = train_data.drop_duplicates()
```

```python
import preprocessor as p

def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    return text
```

```python
train_data['text'] = train_data.apply(preprocess_tweet, axis=1)
```

```python
from gensim.parsing.preprocessing import remove_stopwords

def stopword_removal(row):
    text = row['text']
    text = remove_stopwords(text)
    return text
```

```python
train_data['text'] = train_data.apply(stopword_removal, axis=1)
```

```python
train_data.head()
```

```python
train_data['text'] = train_data['text'].str.lower().str.replace('[^\w\s]',' ').str.replace('\s\s+', ' ')
train_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tweets = train_data['text']
vectorizer = TfidfVectorizer(stop_words='english')

# Learn vocabulary from sentences. 
vectorizer.fit(tweets)

# Get vocabularies.
vectorizer.vocabulary_
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

stopwords = stopwords.words('english')

print(stopwords)

count_vector = CountVectorizer(token_pattern=r'\w{1,}', ngram_range=(1, 2), stop_words = stopwords)
```

```python
from sklearn.model_selection import train_test_split
X = train_data.text
y = train_data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = LogisticRegression()
pipe = Pipeline([
    ('count_vector', CountVectorizer()),  
    ('clf', LogisticRegression()) 
])
pipe.fit(X_train,y_train)
```

```python
from sklearn import metrics
predicted = pipe.predict(X_test)
```

```python
print("accuracy :",  metrics.accuracy_score(predicted, y_test))
```
